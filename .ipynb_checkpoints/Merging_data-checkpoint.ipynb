```python
import pandas as pd
import numpy as np
```

```python
hotels = pd.read_csv("hotels/h_hotels.csv")
hotels.head(5)
```

```python
v_hotels = pd.read_csv("booking-com/hotels.csv")
v_hotels.head(5)
```

```python
df = pd.concat([v_hotels,hotels], axis=0, ignore_index=True)
print(df.head())
```

```python
import uuid
df["hotel_UUID"] = [uuid.uuid4() for _ in range(len(df))]

# Save the updated CSV
df.to_csv("Hotels_data.csv", index=False)
```

```python
df.head(5)
```

```python
prices = pd.read_csv("hotels/h_prices.csv")
prices.head(5)
```

```python
v_prices = pd.read_csv("booking-com/prices.csv")
v_prices.head(5)
```

```python
p_df = pd.concat([prices,v_prices], axis=0, ignore_index=True)
print(p_df.head(5))
```

```python
print(len(df),len(p_df))
```

```python
p_df['hotel_UUID'] = df['hotel_UUID']
```

```python
import uuid
p_df["price_UUID"] = [uuid.uuid4() for _ in range(len(p_df))]

# Save the updated CSV
p_df.to_csv("Price_data.csv", index=False)
```
